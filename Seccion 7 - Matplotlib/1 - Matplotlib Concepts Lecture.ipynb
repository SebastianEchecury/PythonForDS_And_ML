```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
y
```

<h3>FUNCTIONAL</h3>
<p>Single plot</p>

```python
plt.plot(x, y)
plt.xlabel('X Label') # Nombrar eje x
plt.ylabel('Y Label') # Nombrar eje y
plt.title('Title')    # Poner titulo al grafico
```

<p>Multiplots</p>

```python
plt.subplot(1,2,1) # 1 argumento numero de filas, 2 numero de columnas, 3 grafico al que referimos
plt.plot(x,y,'r') # r: red

plt.subplot(1,2,2)
plt.plot(y,x,'b') # b: blue
```

<h3>OBJECT ORIENTED</h3>

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Left, bottom, width, height of the axes

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.4, 0.4, 0.4])#([0.2, 0.2, 0.6, 0.6])

# add_axes seria para fijar margenes. El primer argumento es el margen de la izquierda, el 2 es el inferior, el 3 el derecho y el 4 el superior

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
# fig,axes = plt.subplots()
fig,axes = plt.subplots(nrows=1,ncols=2)
# plt.tight_layout() # para agregar espacio y que se ordene

#for currentAx in axes:
#    currentAx.plot(x,y)

#axes.plot(x,y)

axes[0].plot(x,y) # Puedo poner set_title, xlabel e ylabel
axes[1].plot(y,x)
```

<h3>Figure Size and DPI</h3>

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 5)) # A figsize le paso una tupla con el ancho y la altura

axes.plot(x,x**2, label='X Squared')
axes.plot(x,x**3, label='X Cube')

axes.legend(loc=0) # Para que muestre los labels

plt.tight_layout()
```
