```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
# Muestra por cada 'sex' el promedio de 'total_bill'. X es una categoria, Y es numerico
```

```python
# countplot es como barplot salvo que el estimador explicitamente cuenta el numero de ocurrencias
sns.countplot(x='sex', data=tips)
```

```python
# Estos graficos se utilizan para mostrar la distibucion de datos de categoria
sns.boxplot(x='day',y='total_bill', data=tips) # Le puedo agregar otro argumento (hue) y pasarle otra columna para agrupar informacion
# De abajo para arriba, donde empiezan las cajas son el primer cuartil (Q1) que acumulan el 25% de los datos
# La linea que parte la caja (Q2) es la mediana
# Donde termina la caja es el tercer cuartil (Q3) que acumula el 75% de los datos
```

```python
# Muy similar a boxplot
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True)
```

```python
# Como una combinacion de stripplot y violinplot
sns.swarmplot(x='day',y='total_bill', data=tips)
```

```python
# Se puede combinar swarmplot y violinplot
sns.violinplot(x='day',y='total_bill', data=tips)
sns.swarmplot(x='day',y='total_bill', data=tips, color='black')
```

```python
sns.factorplot(x='day',y='total_bill', data=tips, kind='bar')
```
